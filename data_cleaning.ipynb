```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
mvps.head()
```

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players.head()
```

```python
players["Player"].head(30)
```

```python
#Removing the asteriks in player names
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
#Group players by year for players who played multiple teams in a single year
players.groupby(["Player", "Year"])
```

```python
#Looping through each player group so only one row appears
def single_row(df):
    if df.shape[0]==1:
        return df
    
    else:
        row = df[df["Tm"] == "TOT"]
        #replace TOT with team from last row
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
    
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

```python
#drop the index level 
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
#Combining MVP and Players df
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined.head(5)
```

```python
combined[combined["Pts Won"] > 0]
```

```python
#Fix NaN issues
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined
```

```python
teams = pd.read_csv("teams.csv")
```

```python
teams.head(10)
```

```python
#Remove divisions in df
teams = teams[~teams["W"].str.contains("Division")]
```

```python
#Remove asteriks in team name
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams["Team"].unique()
```

```python
combined["Tm"].unique()
```

```python
#Import nicknames - convert txt to csv
#nicknames = pd.read_csv("nicknames.txt")
#nicknames.to_csv("nicknames.csv", index = None)
```

```python
nicknames = {}

with open("nicknames.csv") as f: 
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head(5)
```

```python
#Merge team df and combined df
stats = combined.merge(teams, how="outer", on=["Team", "Year"])
```

```python
stats
```

```python
del stats["Unnamed: 0"]
```

```python
#configure to correct dtype
stats = stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats.dtypes
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = stats["GB"].str.replace("—", "0")
```

```python
stats["GB"] = pd.to_numeric(stats["GB"])
```

```python
stats.dtypes
```

```python
stats.to_csv("player_mvp_stats.csv")
```
